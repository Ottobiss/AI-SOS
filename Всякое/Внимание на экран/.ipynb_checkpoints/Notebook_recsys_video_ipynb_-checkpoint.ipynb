```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gdown
import gc
import hashlib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

import tensorflow as tf
from tensorflow.keras.layers import Embedding
from transformers import BertTokenizerFast, BertTokenizer, TFBertModel
from catboost import CatBoostClassifier, Pool
```

### Классификация

```python
def create_is_long_view_vectorized(df):
    condition_long = (df['v_duration'] > 300) & (df['watchtime'] > 0.25 * df['v_duration'])
    condition_short = (df['v_duration'] <= 300) & (df['watchtime'] > 30)
    df['is_long_view'] = np.where(condition_long | condition_short, 1, 0)
    return df
```

```python
test = pd.read_parquet('/content/test.parquet')
train = pd.read_parquet('/content/train.parquet')
video_stats = pd.read_parquet('/content/video_stat.parquet')
# test_target = pd.read_parquet('/content/test_target.parquet')
```

```python
test
```

```python
# Мержим train с video_stats
train_merged = pd.merge(train, video_stats, on='video_id', how='inner')

# Добавляем колонку is_long_view
train_merged = create_is_long_view_vectorized(train_merged)
```

```python
# Мержим test с video_stats
test_merged = pd.merge(test, video_stats, on='video_id', how='inner')
test_merged
```

```python
train_merged = train_merged.drop(columns=['watchtime']) # чтоб уровнять с тестом
```

```python
# Готовим таргеты
y_train = train_merged['is_long_view']
train_merged = train_merged.drop(columns=['is_long_view'])
# y_test = test_target
```

```python
# Добавляем метку, чтобы потом разделить трейн и тест обратно
train_merged['dataset_type'] = 'train'
test_merged['dataset_type'] = 'test'

combined_data = pd.concat([train_merged, test_merged], axis=0, ignore_index=True)
```

```python
combined_data['event_timestamp'] = pd.to_datetime(combined_data['event_timestamp'])
combined_data['hour'] = combined_data['event_timestamp'].dt.hour
combined_data['day_of_week'] = combined_data['event_timestamp'].dt.dayofweek
```

```python
# посмотрим на количество каждого признака, чтобы определиться со способом кодирования и размерами баскетов для хэширования

print(combined_data[['user_id']].nunique())
print(combined_data[['region']].nunique())
print(combined_data[['city']].nunique())
print(combined_data[['video_id']].nunique())
print(combined_data[['author_id']].nunique())
```

Обработка айдишников

```python
# Будем хэшировать
def hash_string_to_int(string, num_buckets):
    return int(hashlib.md5(string.encode('utf-8')).hexdigest(), 16) % num_buckets

num_buckets_user_id = 4500000
num_buckets_video_id = 600000
num_buckets_author_id = 120000

combined_data['user_id_hashed'] = combined_data['user_id'].apply(lambda x: hash_string_to_int(x, num_buckets_user_id))
combined_data['video_id_hashed'] = combined_data['video_id'].apply(lambda x: hash_string_to_int(x, num_buckets_video_id))
combined_data['author_id_hashed'] = combined_data['author_id'].apply(lambda x: hash_string_to_int(x, num_buckets_author_id))

user_id_embedding_layer = tf.keras.layers.Embedding(input_dim=num_buckets_user_id, output_dim=16)
video_id_embedding_layer = tf.keras.layers.Embedding(input_dim=num_buckets_video_id, output_dim=16)
author_id_embedding_layer = tf.keras.layers.Embedding(input_dim=num_buckets_author_id, output_dim=8)

user_id_tensor = tf.convert_to_tensor(combined_data['user_id_hashed'], dtype=tf.int32)
video_id_tensor = tf.convert_to_tensor(combined_data['video_id_hashed'], dtype=tf.int32)
author_id_tensor = tf.convert_to_tensor(combined_data['author_id_hashed'], dtype=tf.int32)

user_id_embeddings = user_id_embedding_layer(user_id_tensor).numpy()
video_id_embeddings = video_id_embedding_layer(video_id_tensor).numpy()
author_id_embeddings = author_id_embedding_layer(author_id_tensor).numpy()
```

LabelEncoder для region и city

```python
le_region = LabelEncoder()
le_city = LabelEncoder()

combined_data['region_encoded'] = le_region.fit_transform(combined_data['region'])
combined_data['city_encoded'] = le_city.fit_transform(combined_data['city'])

region_embedding_layer = tf.keras.layers.Embedding(input_dim=combined_data['region_encoded'].nunique(), output_dim=8)
city_embedding_layer = tf.keras.layers.Embedding(input_dim=combined_data['city_encoded'].nunique(), output_dim=8)

region_tensor = tf.convert_to_tensor(combined_data['region_encoded'], dtype=tf.int32)
city_tensor = tf.convert_to_tensor(combined_data['city_encoded'], dtype=tf.int32)

region_embeddings = region_embedding_layer(region_tensor).numpy()
city_embeddings = city_embedding_layer(city_tensor).numpy()
```

LabelEncoder для category_id

```python
le_category = LabelEncoder()
combined_data['category_encoded'] = le_category.fit_transform(combined_data['category_id'])
category_embedding_layer = tf.keras.layers.Embedding(input_dim=combined_data['category_encoded'].nunique(), output_dim=8)
category_tensor = tf.convert_to_tensor(combined_data['category_encoded'], dtype=tf.int32)
category_embeddings = category_embedding_layer(category_tensor).numpy()
```

Собираем вместе

```python
# нормализуем числовые признаки
numerical_features = [
    'v_cr_click_long_view_30_days', 'v_cr_click_long_view_7_days', 'v_duration', 'v_likes', 'v_dislikes', 'v_total_comments',
    'v_year_views', 'v_month_views', 'v_week_views', 'v_day_views', 'v_cr_click_long_view_1_days', 'v_avg_watchtime_1_day'
]
scaler = StandardScaler()
numerical_matrix = scaler.fit_transform(combined_data[numerical_features].values)

user_id_df = pd.DataFrame(user_id_embeddings, columns=[f'user_id_emb_{i}' for i in range(user_id_embeddings.shape[1])])
video_id_df = pd.DataFrame(video_id_embeddings, columns=[f'video_id_emb_{i}' for i in range(video_id_embeddings.shape[1])])
author_id_df = pd.DataFrame(author_id_embeddings, columns=[f'author_id_emb_{i}' for i in range(author_id_embeddings.shape[1])])
category_df = pd.DataFrame(category_embeddings, columns=[f'category_emb_{i}' for i in range(category_embeddings.shape[1])])
region_df = pd.DataFrame(region_embeddings, columns=[f'region_emb_{i}' for i in range(region_embeddings.shape[1])])
city_df = pd.DataFrame(city_embeddings, columns=[f'city_emb_{i}' for i in range(city_embeddings.shape[1])])

numerical_df = pd.DataFrame(numerical_matrix, columns=numerical_features)
final_df = pd.concat([combined_data['dataset_type'], numerical_df, combined_data['hour'], combined_data['day_of_week'], user_id_df, video_id_df, author_id_df, category_df, region_df, city_df], axis=1)
```

```python
# если необходимо, освобождаем память

del combined_data

del video_id_tensor, video_id_embeddings, video_id_embedding_layer, video_id_df
del user_id_tensor, user_id_embeddings, user_id_embedding_layer, user_id_df
del region_df, region_embedding_layer, region_embeddings, region_tensor
del numerical_df, numerical_features, numerical_matrix
del author_id_df, author_id_embedding_layer, author_id_embeddings, author_id_tensor
del category_df, category_embedding_layer, category_embeddings, category_tensor
del city_df, city_embedding_layer, city_embeddings, city_tensor
```

```python
final_df
```

```python
# Разделяем обратно на трейн и тест
X_train = final_df[final_df['dataset_type'] == 'train'].drop(columns=['dataset_type'])
X_test = final_df[final_df['dataset_type'] == 'test'].drop(columns=['dataset_type'])
```

```python
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, random_state=42, stratify=y_train)

# print(f"Размер тренировочной выборки: {X_train.shape}")
# print(f"Размер валидационной выборки: {X_val.shape}")
```

```python
model = CatBoostClassifier(iterations=500, depth=6, learning_rate=0.1, random_seed=42)
model.fit(Pool(X_train, y_train), verbose=100)

test_preds = model.predict(X_test)
test_probs = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, test_preds)
precision = precision_score(y_test, test_preds)
recall = recall_score(y_test, test_preds)
f1 = f1_score(y_test, test_preds)
roc_auc = roc_auc_score(y_test, test_probs)

print(f'Точность (Accuracy): {accuracy:.4f}')
print(f'Точность позитивных предсказаний (Precision): {precision:.4f}')
print(f'Полнота (Recall): {recall:.4f}')
print(f'F1-score: {f1:.4f}')
print(f'ROC-AUC: {roc_auc:.4f}')
```

```python
score_CB = model.get_feature_importance()

importances_CB = pd.DataFrame({
    'feature': X_train.columns,
    'score_CB': score_CB
})

importances_CB = importances_CB.sort_values(by='score_CB', ascending=False)

plt.figure(figsize=(10, 14))
sns.barplot(data=importances_CB, x='score_CB', y='feature')
plt.title('Feature importances')
plt.show()
```

```python
correlation_data = X_test.copy().reset_index(drop=True)
correlation_data['target'] = y_test

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(8, 15))
sns.heatmap(correlation_matrix[['target']].sort_values(by='target', ascending=False), annot=True, cmap='coolwarm')
plt.title("Корреляция признаков теста с целевой переменной")
plt.show()
```
